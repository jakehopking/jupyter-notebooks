Parse the date col into a pd date_time.

```python
unrate['DATE'] = pd.to_datetime(unrate['DATE'])

unrate[:12]
```

```python
plt.plot(unrate.DATE[:12], unrate.VALUE[:12])
plt.xticks(rotation=45)
plt.xlabel("Month")
plt.ylabel("Unemployment Rate")
plt.title("Monthly Unemployment Trends, 1948")
plt.show()
```

```python
fig = plt.figure(figsize=(12, 12))
fig_rows = 6

for i in range(fig_rows):
    ax = fig.add_subplot(fig_rows, 1, i+1)
    start_index = i*12
    end_index = (i+1)*12
    subset = unrate[start_index:end_index]
    ax.plot(subset['DATE'], subset['VALUE'])

plt.show()
```

```python
unrate['MONTH'] = unrate['DATE'].dt.month
```

```python
fig = plt.figure(figsize=(12,3))
plt.plot(unrate['MONTH'][0:12], unrate['VALUE'][0:12], c='red')
plt.plot(unrate['MONTH'][12:24], unrate['VALUE'][12:24], c='blue')
plt.show()
```

```python
fig = plt.figure(figsize=(12, 6))

colours = ["red", "blue", "green", "orange", "white"]
fig_rows = len(colours)

for i in range(fig_rows):
    start_index = i*12
    end_index = (i+1)*12
    subset = unrate[start_index:end_index]
    year = str(1948+i)
    plt.plot(subset['MONTH'], subset['VALUE'], c=colours[i], label=year)

plt.xlabel("Month, Integer")
plt.ylabel("Unemployment Rate, Percent")
plt.title("Monthly Unemployment Trends, 1948-1952")
plt.legend(loc=2)
plt.show()
```
